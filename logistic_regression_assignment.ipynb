**Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose**


**Answer**: Simple Linear Regression is a machine learning method used to find the relationship between one input variable (X) and one output variable (Y).
It draws a straight line that best fits the data.

Purpose: To predict future values

To understand how one value affects another

Example: Predicting salary based on years of experience.


**Question 2: What are the key assumptions of Simple Linear Regression ?**

**Answer:** Simple Linear Regression works properly when these assumptions are true:

There is a linear relationship between X and Y

Errors are random and independent

Variance of errors is constant

There are no extreme outliers

Data is not biased

**Question 3: Write the mathematical equation for a simple linear regression model and explain each term.**

**Answer:**  The equation is:  **y=mx+c**

Where:

y = predicted value

x = input value

m = slope (how much y changes when x changes)

c = intercept (value of y when x = 0)

**Question 5: What is the method of least squares in linear regression?**

**Answer:** The least squares method finds the best straight line by minimizing the total error between actual values and predicted values.

In simple words:
It chooses a line where the difference between real points and line points is minimum.

**Question 6: What is Logistic Regression? How does it differ from Linear Regression?**

**Answer:** Logistic Regression is a machine learning algorithm used when the output is yes or no, true or false, or 0 or 1. It is mainly used for classification problems, such as predicting whether a student will pass or fail or whether an email is spam or not.

Linear Regression is used when the output is a number, like predicting marks, salary, or house price. Logistic Regression does not predict exact numbers; instead, it predicts the probability of an event happening and then converts it into 0 or 1.

Linear Regression uses a straight line, while Logistic Regression uses an S-shaped curve to make decisions. In simple words, Linear Regression is for numeric prediction, and Logistic Regression is for category prediction.

**Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
(Include your Python code and output in the code box below.)**

**Answer: BELOW ðŸ‘‡**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

model = LinearRegression()
model.fit(x, y)

print("Slope:", model.coef_)
print("Intercept:", model.intercept_)
```

**Question 10: How do you interpret the coefficients in a simple linear regression model?**

**Answer:** In a simple linear regression model, there are two main coefficients: slope and intercept.

The slope tells us how much the output value changes when the input value increases by one unit. If the slope is positive, the output increases. If it is negative, the output decreases.

The intercept tells us the value of the output when the input value is zero. It shows the starting point of the line.

In simple words, the slope shows the effect of input on output, and the intercept shows the starting value of the prediction.
